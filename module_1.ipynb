## Handling DICOM data

Implement code which automatically converts the DICOM dataset (3D volume) of the 12 cases of the LCTSC Lung CT Segmentation Challenge into a numpy array  per case

```python
import os
import pydicom

dataset_path = os.path.join(os.getcwd(), 'data', 'Images')

"""
Initilase default dictionary to store the tags
Iterate through patients (cases) and get patient number
"""

for dirpath, dirnames, filenames in os.walk(dataset_path):
    for subdir in dirnames:
        case_id = subdir.split('_')[-1]
        print(f'Case ID: {case_id}')
        subir_path = os.path.join(dataset_path, subdir)
        list_files = [file_name for file_name in os.listdir(subir_path)]
        dicom_file = os.path.join(subir_path, list_files[0])
        metadata = pydicom.dcmread(dicom_file)
```

```python
case_000_dicom = os.path.join(dataset_path, 'Case_000', '1-054.dcm')
metadata = pydicom.dcmread(case_000_dicom)
image_array = metadata.pixel_array

import matplotlib.pyplot as plt

plt.imshow(image_array, cmap='gray')
plt.show()
```
